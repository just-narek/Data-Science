```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart_disease_data.csv')
```

```python
df
```

```python
df.info()
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
df.head()
```

```python
steps = [
    ("imputer", SimpleImputer(strategy='mean')),
     ("scaler", StandardScaler()),
    ('classifier', LogisticRegression())
]
```

```python
pipe = Pipeline(steps)
```

```python
X = df.drop(columns='target')
y = df['target'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
pipe.fit(X_train, y_train)
```

```python
X_train_predection = pipe.predict(X_train)
X_train_accurecy = accuracy_score(y_train, X_train_predection)
print(f"Accurecy for training data is: {X_train_accurecy}")
```

```python
X_test_predection = pipe.predict(X_test)
X_test_accurecy = accuracy_score(y_test, X_test_predection)
print(f"Accurecy for testing data is: {X_test_accurecy}")
```

```python
# Choose your row index
f = 50


# 1. Get the model's prediction
prediction = pipe.predict(X_test.iloc[[f]])[0]

# 2. Get the actual, true answer from y_test
true_answer = y_test.iloc[f]

print(f"Model Predicted: {prediction}")
print(f"Actual Answer:   {true_answer}")

if prediction == true_answer:
    print("✅ The model was right!")
else:
    print("❌ The model was wrong.")
```
